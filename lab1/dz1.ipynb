```python
# Задание №1
class Animal:

    def __init__(self, name, sound):
        self.name = name
        self.sound = sound 
    
    def make_sound(self):
        return self.sound

class Dog(Animal):

    def color(self, color):
        self.color = color
        return self.color

class Cat(Dog):
    pass

Dog.make_sound = 'гав'
Cat.make_sound = 'мяу'

print(Dog.make_sound,
      Cat.make_sound)
```

```python
# Задание №2
class Product:
    def __init__(self, name, 
                 amount, price):
        self.name = name 
        self.amount = amount
        self.price = price
        self.log = []
    
    def amount_raise(self, number): # Увеличение кол-ва товаров
        self.amount += number
        self.log.append(f'Товар: {self.name}, Сумма: {self.price}, Добавили: {number}, Стало штук: {self.amount}.')


    def reducing_amount(self, number): # Уменьшение кол-ва товаров
        if (self.amount - number) > 0:
            self.amount -= number
            self.log.append(f'Товар: {self.name}, Сумма: {self.price}, Продали: {number}, Стало штук: {self.amount}.')
        else:
            return 'Товары закончились'
        
    def final_price(self): # Определение произведения цены на кол-во товаров
         return self.price * self.amount
    
class Warehouse:
    def __init__(self):
        self.goodsList, self.priceList, self.amountList = [], [], []

        
    def add_product(self, prod): # Добавление товара в список 
        self.goodsList.append(prod.name)
        self.priceList.append(prod.price)
        self.amountList.append(prod.amount)

    def delite_product_num(self, prod, del_num_prod): # Удаление товара в конкретном кол-ве
        if prod.name in self.goodsList:
            index = self.goodsList.index(prod.name)
            self.amountList[index] -= del_num_prod
            prod.log.append(f'Товар {prod.name} был удален из общего списка в кол-ве {del_num_prod}')
        else:
            return 'Такого товара нет в списке'
        
    def cost_calc(self): # Расчёт общей стоимости товаров
        self.final_sum = []
        for i in range(len(self.goodsList)):
            self.final_sum.append(self.priceList[i] * self.amountList[i])
        return sum(self.final_sum)
    
class Seller:

    def __init__(self, manager_name):
        self.manager_name = manager_name 
        self.sellingList = []

    def sell(self, sold, number): # Продажа
        if sold.amount - number > 0:
            sold.amount -= number
            income = number * sold.price
            self.sellingList.append(f'{self.manager_name} продал {number} {sold.name} и заработал: {income}, товара осталось: {sold.amount}') 
            sold.log.append(f'{self.manager_name} продал {number} {sold.name} и заработал: {income}, товара осталось: {sold.amount}') 
            warehouse.delite_product_num(sold, number)

        else:
            return 'Слишком много товаров, столько нет на складе'             


# Пример:
# Создание продуктов:
moloko = Product('Молоко', 2000, 50)
kefir = Product('Кефир', 500, 70)
kvas = Product('Квас', 200, 150)

# Проверка методов
moloko.reducing_amount(100)
kefir.amount_raise(150)
print(moloko.amount, kefir.amount, kvas.final_price())

# Создание склада: 
warehouse = Warehouse()

# Проверка методов:
warehouse.add_product(moloko)
warehouse.add_product(kefir)
warehouse.add_product(kvas) 
print(warehouse.cost_calc())
warehouse.delite_product_num(kvas, 20)
print(warehouse.cost_calc())

# Создание продавца: 
Bob = Seller('Боб')

# Проверка методов:
Bob.sell(kefir, 100)
Bob.sell(moloko, 50)
print(Bob.sellingList)

# Проверка логирования: 

print(moloko.log, kefir.log, kvas.log)
print(warehouse.cost_calc())
print(moloko.amount, kefir.amount, kvas.amount,
     moloko.price, kefir.price, kvas.price )


```
